```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, column, desc, lit, expr

spark = SparkSession.builder.appName("chapter5").getOrCreate()

sparkDF = spark.read.format("csv").option("header", "true").load("flight-data/csv/2015-summary.csv")
print(sparkDF.schema)

sparkDF.select(col("DEST_COUNTRY_NAME"), col("ORIGIN_COUNTRY_NAME"), col("count")).show(5)

sparkDF.select(col("count").cast("int")).orderBy(desc("count")).show()

sparkDF.first()

sparkDF.selectExpr("*", "DEST_COUNTRY_NAME=ORIGIN_COUNTRY_NAME as WithinCountry").where("WithinCountry = true").show(10)

sparkDF.selectExpr("avg(count)", "count(distinct(DEST_COUNTRY_NAME))").show()

sparkDF.select("*", lit("1").alias("One")).show(10)

sparkDF.withColumn("WithinCountry", expr("DEST_COUNTRY_NAME==ORIGIN_COUNTRY_NAME")).show(10)

sparkDF.selectExpr("DEST_COUNTRY_NAME", "ORIGIN_COUNTRY_NAME").distinct().show(10)
```
